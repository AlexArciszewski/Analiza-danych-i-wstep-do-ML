```python
X=auto.iloc[:,1:-1]
```

```python
auto = auto.drop('horsepower',axis=1)
```

```python
y=auto.loc[:,'mpg']
```

```python
y
```

```python
X=auto.iloc[:,1:-1]
y=auto.loc[:,'mpg']
```

```python
X.head()
```

```python
X.info()
```

```python
y.head()
```

```python
#Tworzą obiekt modelu regresji liniowej
#Trenują model na danych X i y
#Obliczają współczynnik determinacji R² modelu
```

```python
lr = LinearRegression()
lr.fit(X.to_numpy(),y)
lr.score(X.to_numpy(),y)
```

```python
#Definiują parametry dwóch samochodów do predykcji
#Łączą je w jedną listę
#Używają wytrenowanego modelu do przewidzenia zużycia paliwa (mpg) dla tych samochodów
#Wyświetlają wyniki predykcji
```

```python
my_car1 = [4, 160, 190, 12, 90, 1]
my_car2 = [4, 200, 260, 15, 83, 1]
cars = [my_car1, my_car2]
```

```python
auto_predict =lr.predict(cars)
print(auto_predict)
```
