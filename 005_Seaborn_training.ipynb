```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df =sns.load_dataset('tips')
df.head()
```

```python
sns.scatterplot(data=df, x='total_bill',y='tip')
plt.show()
```

```python
# 1.Modyfikacja rozmiaru
plt.figure(figsize=(15,9))
sns.scatterplot(data=df, x='total_bill',y='tip')
plt.show()
```

```python
# 2 dodanie tytułu
```

```python
plt.figure(figsize=(15,9))
plt.title("Zaleznosc napiwku od wysokosci rachunku")
sns.scatterplot(data=df, x='total_bill',y='tip')
plt.show()
```

```python
# 3 dodanie siatki
```

```python
plt.figure(figsize=(15,9))
plt.title("Zaleznosc napiwku od wysokosci rachunku")
sns.scatterplot(data=df, x='total_bill',y='tip')
plt.grid()
plt.show()
```

```python
# 4 dodanie tekstu na wspolrzednych x,y
```

```python
plt.figure(figsize=(15,9))
plt.title("Zaleznosc napiwku od wysokosci rachunku")
sns.scatterplot(data=df, x='total_bill',y='tip')
plt.grid()
plt.text(30,8,'That point')
plt.show()
```

```python
# Dodanie koloru i rozmiaru tekstu
```

```python
plt.figure(figsize=(15,9))
plt.title("Zaleznosc napiwku od wysokosci rachunku")
sns.scatterplot(data=df, x='total_bill',y='tip')
plt.grid()
plt.text(30,8,'That point', color='purple', size=20)
plt.show()
```

```python
# Zmiana stylu na ggplot
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(15,9))
plt.title("Zaleznosc napiwku od wysokosci rachunku")
sns.scatterplot(data=df, x='total_bill',y='tip')
plt.grid()
plt.text(30,8,'That point', color='purple', size=20)
plt.show()
```

```python
# Przywracanie ustawień z matplotlib
```

```python
plt.rcdefaults()
%matplotlib inline
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(15,9))
plt.title("Zaleznosc napiwku od wysokosci rachunku")
sns.scatterplot(data=df, x='total_bill',y='tip')
plt.grid()
plt.text(30,8,'That point', color='purple', size=20)
plt.show()
```

```python
# Modyfikacja kontekstu context czyli grubość liter nagłówka osi itd..kolorystyki nie zmienia
```

```python
sns.set_style('darkgrid')
sns.set_context('talk')
plt.figure(figsize=(15,9))
plt.title("Zaleznosc napiwku od wysokosci rachunku")
sns.scatterplot(data=df, x='total_bill',y='tip')
plt.grid()
plt.text(30,8,'That point', color='purple', size=20)
plt.show()
```

```python
sns.set_style('darkgrid')
sns.set_context('poster')
plt.figure(figsize=(15,9))
plt.title("Zaleznosc napiwku od wysokosci rachunku")
sns.scatterplot(data=df, x='total_bill',y='tip')
plt.grid()
plt.text(30,8,'That point', color='purple', size=20)
plt.show()
```
